```python
import pandas as pd
df=pd.read_csv("Downloads/Customer Call List.csv",encoding="utf-8")
df.head()
```

```python
df.columns
```

```python
df = df.drop_duplicates()
df
```

```python
import re
# Assuming df['Phone_Number'] contains strings representing phone numbers
df['Phone_Number'] = df['Phone_Number'].astype(str)
phone_number_pattern = re.compile(r'^\D*(\d{3})\D*(\d{3})\D*(\d{4})\D*$')
# Function to extract and reformat phone numbers
def reformat_phone_number(phone):
    match = phone_number_pattern.match(phone)
    if match:
        return f'({match.group(1)}) {match.group(2)}-{match.group(3)}'
    else:
        return None
df['Phone_Number'] = df['Phone_Number'].apply(reformat_phone_number)
df
```

```python
df = df.drop(['Unnamed: 8', 'Unnamed: 9'], axis=1)
df
```

```python
df['Phone_Number'].fillna('Not Available', inplace=True)
df
```

```python
df.isnull().any()
```

```python
df[df['Last_Name'].isna()==1]
```

```python
df['Last_Name']=df['Last_Name'].fillna("Not Defined")
df
```

```python
df[df['Do_Not_Contact'].isna()==1]
```

```python
df['Do_Not_Contact']=df['Do_Not_Contact'].fillna("no status")
df
```

```python
df.drop('Not_Useful_Column', axis=1, inplace=True)
df
```

```python
# Replace specific characters in the 'Last_Name' column
df['Last_Name'] = df['Last_Name'].str.replace(r'[.../-]', '', regex=True)
df
```

```python
# Replace 'N' with 'No' and 'Y' with 'Yes' for 'Paying Customer'
df['Paying Customer'] = df['Paying Customer'].replace({'N': 'No', 'Y': 'Yes'})
df['Paying Customer'] = df['Paying Customer'].replace({'N/a': 'Not Available'})

# Replace 'N/a' with 'Not Available' for 'Address'
df['Address'] = df['Address'].replace({'N/a': 'Not Available'})

# Replace 'N' with 'No' and 'Y' with 'Yes' for 'Do_Not_Contact'
df['Do_Not_Contact'] = df['Do_Not_Contact'].replace({'N': 'No', 'Y': 'Yes'})
df
```
